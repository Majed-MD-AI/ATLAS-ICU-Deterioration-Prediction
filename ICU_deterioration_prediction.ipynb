## ICU Deterioration Prediction

This project aims to predict early signs of patient deterioration in ICU using simple clinical features such as age,heart rate,and blood pressure.

## Clinical Importance
Early detection of patient deterioration in ICU can significantly reduce mortality and improve outcomes.
This project demonstrates a simple machine learning approach to identify high-risk patients using basic vital signs.

## Model Comparison
Two machine learning models were used:
- Logistic Regression: A simple and interpretable model for binary classification.
- Random Forest: A more complex model that can capture non-linear relationships.
- The performance of both models was evaluated using accuracy.

## Implementation
This section demonstrates the step-by-step implementation of a machine learning pipeline for predicting ICU patient deterioration.
The workflow includes data preparation, model training using Logistic Regression and Random Forest, and performance evaluation using accuracy.

### Data Preparation 

```python
import pandas as pd
```

```python
data = {
        "Age": [65, 72, 50, 80, 60] ,
        "HeartRate": [110, 95, 88, 120, 100] ,
        "BloodPressure": [90, 100, 110, 85, 95] ,
        "Outcome": [1, 0, 0, 1, 1]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

### Model Training 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = df[["Age", "HeartRate","BloodPressure"]]
Y = df["Outcome"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
model.predict(X_test)
```

```python
model.predict([[70, 115, 85]])
```

```python
import pandas as pd
new_patient = pd.DataFrame({
    "Age": [70],
    "HeartRate": [115],
    "BloodPressure": [85]
})
model.predict(New_patient)
    
```

```python
import pandas as pd
new_patient = pd.DataFrame({
    "Age": [70],
    "HeartRate": [115],
    "BloodPressure": [85]
})
model.predict(new_patient)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, Y_train)
rf_model.predict(new_patient)
```

```python
print("Logistic Regression Prediction:", model.predict(new_patient))
print("Random Forest Prediction:", rf_model.predict(new_patient)
```

```python
print("Logistic Regression Prediction:", model.predict(new_patient))
print("Random Forest Prediction:", rf_model.predict(new_patient))
```

### Evaluation 

```python
from sklearn.metrics import accuracy_score
log_pred = model.predict(X_test)
rf_pred = rf_model.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(Y_test, log_pred))
print("Random Forest Accuracy:", accuracy_score(Y_test, rf_pred))
```
